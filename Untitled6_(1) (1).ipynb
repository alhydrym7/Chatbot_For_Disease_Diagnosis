```python

def tokenize(sentence):
    if isinstance(sentence, str):
        return nltk.word_tokenize(sentence)
    else:
        return nltk.word_tokenize(str(sentence))
def stem(word):
    return stemmer.stem(word.lower())

def bag_of_words(tokenized_sentence, words):

    sentence_words = [stem(word) for word in tokenized_sentence]
    bag = np.zeros(len(words), dtype=np.float32)
    for idx, w in enumerate(words):
        if w in sentence_words:
            bag[idx] = 1

    return bag
```

```python
"""Data sterilization"""

def delete_links(input_text):
    pettern  = r'''(?i)\b((?:https?://|www\d{0,3}[.]|[a-z0-9.\-]+[.][a-z]{2,4}/)(?:[^\s()<>]+|\(([^\s()<>]+|(\([^\s()<>]+\)))*\))+(?:\(([^\s()<>]+|(\([^\s()<>]+\)))*\)|[^\s`!()\[\]{};:'".,<>?«»“”‘’]))'''
    out_text = re.sub(pettern, ' ', input_text)
    return out_text

"""Fixing word lengthening"""

def delete_repeated_characters(input_text):
    pattern  = r'(.)\1{2,}'
    out_text = re.sub(pattern, r"\1\1", input_text)
    return out_text

"""Delete bad symbols"""

def clean_text(input_text):
    replace = r'[/(){}\[\]|@âÂ,;\?\'\"\*…؟–’،!&\+-:؛-]'
    out_text = re.sub(replace, " ", input_text)
    words = nltk.word_tokenize(out_text)
    words = [word for word in words if word.isalpha()]
    out_text = ' '.join(words)
    return out_text

"""Delete stopwords"""

def delete_stopwords(input_text):
    stop_words = set(nltk.corpus.stopwords.words("english"))
    tokenizer = nltk.tokenize.WhitespaceTokenizer()
    tokens = tokenizer.tokenize(input_text)
    wnl = nltk.WordNetLemmatizer()
    lemmatizedTokens =[wnl.lemmatize(t) for t in tokens]
    out_text = [w for w in lemmatizedTokens if not w in stop_words]
    out_text = ' '.join(out_text)
    return out_text
```

```python
def text_prepare(input_text):
    out_text = delete_links(input_text)
    out_text = delete_repeated_characters(out_text)
    out_text = clean_text(out_text)
    out_text = delete_stopwords(out_text)
    out_text = out_text.lower()
    return out_text
```

```python
import torch
import torch.nn as nn
class NeuralNet(nn.Module):
    def __init__(self, input_size, hidden_size, num_classes):
        super(NeuralNet, self).__init__()
        self.l1 = nn.Linear(input_size, hidden_size)
        self.l2 = nn.Linear(hidden_size, hidden_size)
        self.l3 = nn.Linear(hidden_size, num_classes)
        self.relu = nn.ReLU()
        self.Softmax = nn.Softmax()

    def forward(self, x):
        out = self.l1(x)
        out = self.relu(out)
        out = self.l2(out)
        out = self.relu(out)
        out = self.l3(out)
        out = self.Softmax(out)
        return out


# class TextClassificationModel(nn.Module):
#     def __init__(self, vocab_size, embedding_dim, output_dim):
#         super(TextClassificationModel, self).__init__()
#         self.embedding = nn.Embedding(vocab_size, embedding_dim)
#         self.conv1d = nn.Conv1d(in_channels=embedding_dim, out_channels=128, kernel_size=3)
#         self.global_pooling = nn.AdaptiveAvgPool1d(1)
#         self.fc = nn.Linear(128, output_dim)

#     def forward(self, text):
#         embedded = self.embedding(text)
#         embedded = embedded.permute(0, 2, 1)  # بما أننا نستخدم Conv1d، يجب تغيير ترتيب الأبعاد
#         conv_out = F.relu(self.conv1d(embedded))
#         global_pool = self.global_pooling(conv_out).squeeze(2)
#         output = self.fc(global_pool)
#         return output
```

```python
import numpy as np
import random
import json

import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader

```

Reading Data From a JSON File

```python
with open('output_file_9.json', 'r', encoding="utf-8") as json_data:
    training_data = json.load(json_data)
```

**Data Preparation**

```python
all_words = []
tags = []
xy = []
for intent in training_data['intents']:
    tag = intent['tag']
    tags.append(tag)
    for response in intent['responses']:
        text_clen = text_prepare(response)
        w = tokenize(text_clen)
        all_words.extend(w)
        xy.append((w, tag))


ignore_words = ['?', '.', '!']
all_words = [stem(w) for w in all_words if w not in ignore_words]
all_words = sorted(set(all_words))
tags = sorted(set(tags))

print(len(xy), "responses")
print(len(tags), "tags:", tags)
print(len(all_words), "unique stemmed words:", all_words)
```

**Preparing Data For Training**

```python
X_train = []
y_train = []
for (pattern_sentence, tag) in xy:
    bag = bag_of_words(pattern_sentence, all_words)
    X_train.append(bag)
    label = tags.index(tag)
    y_train.append(label)

X_train = np.array(X_train)
y_train = np.array(y_train)
```

**Define Training Parameters**

```python
num_epochs = 500
batch_size = 32
learning_rate = 0.001
input_size = len(X_train[0])
hidden_size = 64
output_size = len(tags)
print(input_size, output_size)
```

**Create A ChatDataset Class**

```python
class ChatDataset(Dataset):

    def __init__(self):
        self.n_samples = len(X_train)
        self.x_data = X_train
        self.y_data = y_train

    def __getitem__(self, index):
        return self.x_data[index], self.y_data[index]

    def __len__(self):
        return self.n_samples

dataset = ChatDataset()
train_loader = DataLoader(dataset=dataset,
                          batch_size=batch_size,
                          shuffle=True,
                          num_workers=0)
```

**Create A Neural Network Model**

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = TextClassificationModel(len(all_words), 128, output_size).to(device)
```

**Definition Of Loss Function And Optimizer Of The Model**

```python
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)
```

**Train The Model**

```python
for epoch in range(num_epochs):
    total = 0
    correct = 0
    running_loss = 0.0

    for (words, labels) in train_loader:
        words = words.to(device)
        labels = labels.to(dtype=torch.long).to(device)

        # Forward pass
        outputs = model(words)
        loss = criterion(outputs, labels)

        # Backward and optimize
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        running_loss += loss.item() * words.size(0)
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

    # Compute epoch accuracy and loss
    epoch_loss = running_loss / len(train_loader.dataset)
    epoch_accuracy = 100 * correct / total

    if (epoch + 1) % 100 == 0:
        print(f'Epoch [{epoch + 1}/{num_epochs}], Accuracy: {epoch_accuracy:.2f}%, Loss: {epoch_loss:.4f}')

print(f'Final Accuracy: {epoch_accuracy:.2f}%')
print(f'Final Loss: {epoch_loss:.4f}')
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torch.nn.utils.rnn import pad_sequence
from torch.utils.data import DataLoader, TensorDataset
import matplotlib.pyplot as plt

# قراءة ملف البيانات من CSV (يجب استبدال 'path_to_csv_file' بمسار الملف الفعلي)
# data = pd.read_csv("/content/Symptom2Disease.csv")

# استعراض بعض الصفوف الأولى من البيانات
data.head()

# عرض القيم الموجودة في عمود "label"
data["label"].unique()

# تحويل الفئات إلى أرقام باستخدام LabelEncoder
label_encoder = LabelEncoder()
data['label'] = label_encoder.fit_transform(data['label'])

# تجهيز البيانات للتدريب والاختبار
X_train, X_test, y_train, y_test = train_test_split(data['text'], data['label'], test_size=0.1, random_state=42)

# تحويل النصوص إلى تنسور تشير أوبجكت باستخدام map
def tokenize_text(texts):
    tokenized_texts = []
    for text in texts:
        tokens = [token.lower() for token in text.split()]
        tokenized_texts.append(tokens)
    return tokenized_texts

X_train_tokenized = tokenize_text(X_train)
X_test_tokenized = tokenize_text(X_test)

# بناء قاموس الكلمات
word_to_index = {}
for tokens in X_train_tokenized + X_test_tokenized:
    for token in tokens:
        if token not in word_to_index:
            word_to_index[token] = len(word_to_index)

# تحويل النصوص إلى مصفوفة رقمية
def texts_to_padded_sequences(texts, word_to_index, max_length):
    sequences = [torch.tensor([word_to_index[token] for token in tokens], dtype=torch.long) for tokens in texts]
    padded_sequences = pad_sequence(sequences, batch_first=True, padding_value=0)
    padded_sequences = nn.functional.pad(padded_sequences, (0, max_length - padded_sequences.size(1)))
    return padded_sequences

max_length = 45
X_train_padded = texts_to_padded_sequences(X_train_tokenized, word_to_index, max_length)
X_test_padded = texts_to_padded_sequences(X_test_tokenized, word_to_index, max_length)
y_train_tensor = torch.tensor(y_train.values, dtype=torch.long)
y_test_tensor = torch.tensor(y_test.values, dtype=torch.long)

# حساب عدد الفئات
num_of_classes = len(label_encoder.classes_)

class TextClassificationModel(nn.Module):
    def __init__(self, vocab_size, embedding_dim, output_dim):
        super(TextClassificationModel, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim)
        self.conv1d = nn.Conv1d(in_channels=embedding_dim, out_channels=128, kernel_size=3)
        self.global_pooling = nn.AdaptiveAvgPool1d(1)
        self.fc = nn.Linear(128, output_dim)

    def forward(self, text):
        embedded = self.embedding(text)
        embedded = embedded.permute(0, 2, 1)  # بما أننا نستخدم Conv1d، يجب تغيير ترتيب الأبعاد
        conv_out = F.relu(self.conv1d(embedded))
        global_pool = self.global_pooling(conv_out).squeeze(2)
        output = self.fc(global_pool)
        return output

vocab_size = len(word_to_index)
embedding_dim = 128
output_dim = num_of_classes

# إنشاء نموذج الشبكة العصبية في PyTorch
model = TextClassificationModel(vocab_size, embedding_dim, output_dim)

# تحديد دالة الخسارة والمحسن
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters())

# Convert the tensors into a dataset
train_dataset = TensorDataset(X_train_padded, y_train_tensor)
test_dataset = TensorDataset(X_test_padded, y_test_tensor)

# Create data loaders
batch_size = 64  # You can adjust the batch size based on your memory capacity
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

# التدريب
# قائمة لتخزين القيم المحسوبة خلال التدريب
train_losses = []
test_losses = []
test_accuracies = []

# التدريب
num_epochs = 500
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

for epoch in range(num_epochs):
    model.train()
    for batch_text, batch_labels in train_loader:
        batch_text, batch_labels = batch_text.to(device), batch_labels.to(device)
        optimizer.zero_grad()
        output = model(batch_text)
        loss = criterion(output, batch_labels)
        loss.backward()
        optimizer.step()

    # اختبار النموذج على البيانات الاختبار
    model.eval()
    with torch.no_grad():
        test_loss = 0
        correct = 0
        total = 0
        for batch_text, batch_labels in test_loader:
            batch_text, batch_labels = batch_text.to(device), batch_labels.to(device)
            output = model(batch_text)
            test_loss += criterion(output, batch_labels).item()
            _, predicted = output.max(1)
            total += batch_labels.size(0)
            correct += (predicted == batch_labels).sum().item()

    # حساب المتوسطات على البيانات الاختبار
    test_loss /= len(test_loader)
    test_accuracy = correct / total

    # حفظ القيم في القوائم
    train_losses.append(loss.item())
    test_losses.append(test_loss)
    test_accuracies.append(test_accuracy)

    # طباعة نتائج الاختبار
    print(f"Epoch {epoch + 1}/{num_epochs}, Test Loss: {test_loss}, Test Accuracy: {test_accuracy}")

# رسم الخطأ والدقة والفال اكيروسي والفال لوس
plt.figure(figsize=(10, 5))
plt.plot(range(1, num_epochs + 1), train_losses, label='Train Loss')
plt.plot(range(1, num_epochs + 1), test_losses, label='Test Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(range(1, num_epochs + 1), test_accuracies, label='Test Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torch.nn.utils.rnn import pad_sequence
from torch.utils.data import DataLoader, TensorDataset
import matplotlib.pyplot as plt

# قراءة ملف البيانات من CSV (يجب استبدال 'path_to_csv_file' بمسار الملف الفعلي)
# data = pd.read_csv("/content/Symptom2Disease.csv")

# استعراض بعض الصفوف الأولى من البيانات
data.head()

# عرض القيم الموجودة في عمود "label"
data["answer"].unique()

# تحويل الفئات إلى أرقام باستخدام LabelEncoder
label_encoder = LabelEncoder()
data['answer'] = label_encoder.fit_transform(data['answer'])

# تجهيز البيانات للتدريب والاختبار
X_train, X_test, y_train, y_test = train_test_split(data['question'], data['answer'], test_size=0.1, random_state=42)

# تحويل النصوص إلى تنسور تشير أوبجكت باستخدام map
def tokenize_text(texts):
    tokenized_texts = []
    for text in texts:
        tokens = [token.lower() for token in text.split()]
        tokenized_texts.append(tokens)
    return tokenized_texts

X_train_tokenized = tokenize_text(X_train)
X_test_tokenized = tokenize_text(X_test)

# بناء قاموس الكلمات
word_to_index = {}
for tokens in X_train_tokenized + X_test_tokenized:
    for token in tokens:
        if token not in word_to_index:
            word_to_index[token] = len(word_to_index)

# تحويل النصوص إلى مصفوفة رقمية
def texts_to_padded_sequences(texts, word_to_index, max_length):
    sequences = [torch.tensor([word_to_index[token] for token in tokens], dtype=torch.long) for tokens in texts]
    padded_sequences = pad_sequence(sequences, batch_first=True, padding_value=0)
    padded_sequences = nn.functional.pad(padded_sequences, (0, max_length - padded_sequences.size(1)))
    return padded_sequences

max_length = 45
X_train_padded = texts_to_padded_sequences(X_train_tokenized, word_to_index, max_length)
X_test_padded = texts_to_padded_sequences(X_test_tokenized, word_to_index, max_length)
y_train_tensor = torch.tensor(y_train.values, dtype=torch.long)
y_test_tensor = torch.tensor(y_test.values, dtype=torch.long)

# حساب عدد الفئات
num_of_classes = len(label_encoder.classes_)

def adjust_learning_rate(optimizer, epoch):
    lr = 0.001
    if epoch > 10:
        lr = 0.0001
    for param_group in optimizer.param_groups:
        param_group['lr'] = lr

# إنشاء نموذج الشبكة العصبية بالطبقة LSTM
class LSTMTextClassificationModel(nn.Module):
    def __init__(self, vocab_size, embedding_dim, hidden_dim, output_dim):
        super(LSTMTextClassificationModel, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim)
        self.lstm = nn.LSTM(embedding_dim, hidden_dim, batch_first=True, dropout=0.2)  # إضافة dropout
        self.fc = nn.Linear(hidden_dim, output_dim)

    def forward(self, text):
        embedded = self.embedding(text)
        lstm_out, _ = self.lstm(embedded)
        lstm_out = lstm_out[:, -1, :]  # Using the last output in the sequence
        output = self.fc(lstm_out)
        return output

vocab_size = len(word_to_index)
embedding_dim = 128
hidden_dim = 64
output_dim = num_of_classes

# إنشاء نموذج الشبكة العصبية بالطبقة LSTM
model = LSTMTextClassificationModel(vocab_size, embedding_dim, hidden_dim, output_dim)

# تحديد دالة الخسارة والمحسن
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)  # قيمة learning rate مُحدّدة هنا

# Convert the tensors into a dataset
train_dataset = TensorDataset(X_train_padded, y_train_tensor)
test_dataset = TensorDataset(X_test_padded, y_test_tensor)

# Create data loaders
batch_size = 64  # You can adjust the batch size based on your memory capacity
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

# التدريب
# قائمة لتخزين القيم المحسوبة خلال التدريب
train_losses = []
test_losses = []
test_accuracies = []

# التدريب
num_epochs = 500
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

for epoch in range(num_epochs):
    model.train()
    for batch_text, batch_labels in train_loader:
        batch_text, batch_labels = batch_text.to(device), batch_labels.to(device)
        optimizer.zero_grad()
        output = model(batch_text)
        loss = criterion(output, batch_labels)
        loss.backward()
        optimizer.step()

    # اختبار النموذج على البيانات الاختبار
    model.eval()
    with torch.no_grad():
        test_loss = 0
        correct = 0
        total = 0
        for batch_text, batch_labels in test_loader:
            batch_text, batch_labels = batch_text.to(device), batch_labels.to(device)
            output = model(batch_text)
            test_loss += criterion(output, batch_labels).item()
            _, predicted = output.max(1)
            total += batch_labels.size(0)
            correct += (predicted == batch_labels).sum().item()

    # حساب المتوسطات على البيانات الاختبار
    test_loss /= len(test_loader)
    test_accuracy = correct / total

    # حفظ القيم في القوائم
    train_losses.append(loss.item())
    test_losses.append(test_loss)
    test_accuracies.append(test_accuracy)

    # تعديل معدل التعلم
    adjust_learning_rate(optimizer, epoch)

    # طباعة نتائج الاختبار
    print(f"Epoch {epoch + 1}/{num_epochs}, Test Loss: {test_loss}, Test Accuracy: {test_accuracy}")


# رسم الخطأ والدقة والفال اكيروسي والفال لوس
plt.figure(figsize=(10, 5))
plt.plot(range(1, num_epochs + 1), train_losses, label='Train Loss')
plt.plot(range(1, num_epochs + 1), test_losses, label='Test Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(range(1, num_epochs + 1), test_accuracies, label='Test Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
